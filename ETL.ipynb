# ETL

Vamos a hacer un archivo ETL para asi poder trabajar con más facilidad con los datasets

```python
import pandas as pd
import numpy as np
```

Cargaremos los datasets

```python
Hechos = pd.read_csv('data/homicidios_HECHOS.csv')
Victimas = pd.read_csv('data/homicidios_VICTIMAS.csv')
```

### Proceso ETL del dataset de Hechos

```python
Hechos.head()
```

```python
Victimas.head()
```

```python
Hechos.columns = Hechos.columns.str.lower().str.replace(' ', '_').str.replace('á', 'a').str.replace('é', 'e').str.replace('í', 'i').str.replace('ó', 'o').str.replace('ú', 'u')
Hechos.rename(columns={'pos_x': 'longitud', 'pos_y': 'latitud'}, inplace = True)
Hechos
```

Normalizaremos el tipo de dato de la columna "fecha"

```python
# Normalización de la columna fecha
Hechos['fecha'] = pd.to_datetime(Hechos['fecha'])
```

```python
Hechos.dtypes
```

Eliminaremos columnas no relevantes para nuestro analisis

```python
Hechos.drop(columns=['lugar_del_hecho','calle','altura','cruce','direccion_normalizada','xy_(caba)', 'aaaa','dd'],inplace=True)  
```

Verificamos si nos quedo algún valor nulo en nuestro dataset

```python
Hechos.isnull().sum()
```

Buscamos duplicados

```python
Hechos[Hechos.duplicated()]
```

```python
Hechos.head()
```

```python
Hechos.to_csv('datasets/homicidios_hechos.csv', index=False)
```

### Proceso ETL Para el dataset victimas

```python
Victimas
```

Normalizaremos el nombre de las columnas

```python
# Normalizamos el nombre de las columnas
Victimas.rename(columns=str.lower, inplace=True)
Victimas.head()
```

Normalizamos los datos de la columna fecha a datetime

```python
# Normalización de la columna fecha
Victimas['fecha'] = pd.to_datetime(Victimas['fecha'])
```

Eliminaremos las columnas irrelevantes para nuesto análisis

```python
# Eliminación de columnas
Victimas.drop(columns=['aaaa', 'mm', 'dd', 'fecha_fallecimiento'], inplace=True)
```

Verificamos si existen valores nulos

```python
# Busqueda de nulos
Victimas.isnull().sum()
```

Buscamos duplicados

```python
Victimas[Victimas.duplicated()]
```

Exportamos el dataset limpio

```python
Victimas
```

```python
Victimas.to_csv('datasets/homicidios_victimas.csv')
```
